# Bubble Sort
Sort an array using Bubble sort


## Scenario 1 - NORMAL ARRAY
Sorting a simple 1D array in alphabetical order. Consider the array called **names** with the following elements

|  |  |
| ---    | --- |
|**1** \|   | John | 
|**2** \|   | Bob | 
| **3** \| | Mary |
| **4** \| | Sally | 
| **5** \| | Andy |

````python
names = ["John", "Bob", "Mary", "Sally", "Andy"]

def sort():
    long = len(names)
    top = long - 1
    sorted = False
    while not sorted:
        sorted = True
        for k in range(top):
            if names[k] > names[k+1]:
                sorted = False
                temp = names[k]
                names[k] = names[k+1]
                names[k+1] = temp
        top = top -1
        
#MAIN
print(names)     # Before sort
sort() 
print(names)     # After sort
````

### Scenario 1 Pseudocode:
```
DECLARE names: ARRAY[1:10] OF string     // assume we have 10 names in the array 

PROCEDURE sort()
    DECLARE k, top, long: INTEGER
    DECLARE sorted : BOOLEAN
    DECLARE temp : STRING
    
    long ← 10
    top ← long - 1
    sorted ← FALSE
    WHILE NOT sorted DO
        sorted ← TRUE
        FOR k ← 1 TO top
            IF names[k] > names[k+1]
                THEN
                    sorted ← False
                    temp ← names[k]
                    names[k] ← names[k+1]
                    names[k+1] ← temp
           ENDIF
        ENDFOR
        top ← top -1
     ENDWHILE
ENDPROCEDURE
```

### Scenario 2 - ARRAY OF TYPE RECORD
Sorting an 1D array of type record according of one of the fields. 

Consider the array called **students.** At each index in the array, a student’s name and age are stored as the fields of the record. Sort this array in ascending order of age.

|  |  |
| ---    | --- |
|**1** \|   | **name:** : John <br> **age:** 24 | 
|**2** \|   | **name:** : Bob <br> **age:** 12  | 
| **3** \| | **name:** : Mary <br> **age:** 77  |
| **4** \| | **name:** : Sally <br> **age:** 19  | 
| **5** \| | **name:** : Andy <br> **age:** 4  |


````python
class person:
        name = None
        age = 0

students = [person() for j in range(5)]
students[0].name = "John"; students[1].name = "Bob"; students[2].name = "Mary"; 
students[3].name = "Sally"; students[4].name = "Andy"
students[0].age = 24; students[1].age = 12; students[2].age = 77; 
students[3].age = 19; students[4].age = 4


def sort():
    long = len(students)
    top = long - 1
    sorted = False
    while not sorted:
        sorted = True
        for k in range(top):
            if students[k].age > students[k+1].age:
                sorted = False
                temp = students[k]
                students[k] = students[k+1]
                students[k+1] = temp
        top = top -1

def display():
    long = len(students)
    for j in range(long):
        print(students[j].name ,students[j].age)


# #MAIN
display()     # Before sort
sort()
print("\nAFTER SORT")
display()     # After sort
````

### Scenario 2 Pseudocode

```
    DECLARE name : STRING

ENDTYPE

DECLARE students : ARRAY[1:10] OF person      // assume we have 10 persons in the array

    DECLARE temp : person

    top ← long - 1
    sorted ← False

        sorted ← True

                    students[k+1] ← temp
             ENDIF  
        ENDFOR

      ENDWHILE
ENDPROCEDURE
```

### Scenario 3 - 2D ARRAY
Sorting a 2D array into ascending order according to one of the columns. Consider an 2D array called **kids**. Data of 10 students are stored. The first column stores the name of the student, the second the surname and the third the name of their favourite pet.

|  |  |  |  |
| ---    | --- | --- | --- |
|**1** \|   | John | Simons | Fluffy |
|**2** \|   | Bob | Jones | Butch |
| **3** \| | Mary | Adams | Snowy |
| **4** \| | Sally | Williams | Jelly Bean |
| **5** \| | Andy | Macdonald | Space |


````python
kids = [["John", "Simons", "Fluffy"], 
            ["Bob", "Jones", "Butch"], 
            ["Mary", "Adams", "Snowy"], 
            ["Sally", "Williams", "Jelly Bean"], 
            ["Andy", "Macdonald", "Space"]]

def sort():
    long = len(kids)
    top = long - 1
    sorted = False
    while not sorted:
        sorted = True
        for k in range(top):
            if kids[k][1] > kids[k+1][1]:   #Compare the surname column
                sorted = False
                
                #Swap names
                temp = kids[k][0]
                kids[k][0] = kids[k+1][0]
                kids[k+1][0] = temp
                
                #Swap surname
                temp = kids[k][1]
                kids[k][1] = kids[k+1][1]
                kids[k+1][1] = temp
                
                #Swap pets
                temp = kids[k][2]
                kids[k][2] = kids[k+1][2]
                kids[k+1][2] = temp
        top = top -1
        
def display():
    for r in range(len(kids)):
        for c in range(3):
            print(kids[r][c], end=", ")
        print()
        
#MAIN
display()     # Before sort
sort()
print("\nAFTER SORT")
display()     # After sort
````

### Scenario 3 Pseudocode
```

    sorted ← FALSE

        sorted ← TRUE

                sorted ← FALSE
                
                //Swap names

                kids[k, 1] ← kids[k+1, 1]
                kids[k+1, 1] = temp
                
                //Swap surname
                temp ← kids[k, 2]
                kids[k, 2] ← kids[k+1, 2]
                kids[k+1, 2] ← temp
                
                //Swap pets
                temp ← kids[k, 3]
                kids[k, 3] ← kids[k+1, 3]
                kids[k+1, 3] ← temp
            ENDIF
        ENDFOR
        top = top -1
     ENDWHILE
ENDPROCEDURE
```
